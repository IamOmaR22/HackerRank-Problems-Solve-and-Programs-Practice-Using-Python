```python
import pandas as pd
r=15
x=0
wd=0
y=r
dic={}
k=-1
if isinstance(r, int):
    p=1-r
else:
    p=(5/4)-r
while x<=y:
    k=k+1
    if p>=0:
        e=p
        x=x+1
        y=y-1
        a=x
        b=y
        f=2*x
        g=2*y
        p=p+f+1-g
        c=0
        keys = k
        dic[keys] = e,a,b,c,p
    else:
        e=p
        x=x+1
        y=y
        a=x
        b=y
        f=2*x
        p=p+f+1
        c=0
        keys = k
        dic[keys] = e,a,b,p,c
df = pd.DataFrame(dic.values(), columns=["P","X", "Y","Pk1","Pk2"]) 
df        
```
